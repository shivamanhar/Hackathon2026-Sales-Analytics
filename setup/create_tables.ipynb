#### Script for create silver table

```python
%sql
DROP TABLE IF EXISTS silver_orders;
CREATE TABLE IF NOT EXISTS silver_orders (
  order_id STRING,
  order_date DATE,
  customer_name STRING,
  state STRING,
  city STRING,
  _rescued_data STRING
)
```

```python
%sql
DROP TABLE IF EXISTS silver_order_details;
CREATE TABLE IF NOT EXISTS silver_order_details (  
  surrogate_key STRING,
  order_id STRING,
  amount  DECIMAL(18,2),
  profit  DECIMAL(18,2),
  quantity INT,
  category STRING,
  sub_category STRING,
  _rescued_data STRING
)
```

```python
%sql
DROP TABLE IF EXISTS silver_return_orders;
CREATE TABLE IF NOT EXISTS silver_return_orders (
  surrogate_key STRING,
  order_id STRING,
  amount  DECIMAL(18,2),
  profit  DECIMAL(18,2),
  quantity INT,
  category STRING,
  sub_category STRING
)
```

```python
%sql
DROP TABLE IF EXISTS silver_actual_orders;
CREATE TABLE IF NOT EXISTS silver_actual_orders (
  surrogate_key STRING,
  order_id STRING,
  amount  DECIMAL(18,2),
  profit  DECIMAL(18,2),
  quantity INT,
  category STRING,
  sub_category STRING
)
```

```python
%sql
DROP TABLE IF EXISTS silver_sales_target;
CREATE TABLE IF NOT EXISTS silver_sales_target (
  surrogate_key STRING,
  month_of_order_date DATE,
  category STRING,
  target DECIMAL(18,2),
  _rescued_data STRING
)
```

```python
%sql
DROP TABLE IF EXISTS gold_sales_facts;
CREATE TABLE IF NOT EXISTS gold_sales_facts (
  order_id STRING,
  order_date DATE,
  customer_name STRING,
  state STRING,
  city STRING,
  amount  DECIMAL(18,2),
  profit  DECIMAL(18,2),
  quantity INT,
  category STRING,
  sub_category STRING,
  insert_timestamp TIMESTAMP,
  update_timestamp TIMESTAMP
)
```

```python
%sql
DROP TABLE IF EXISTS gold_daily_order_metrics;
CREATE TABLE IF NOT EXISTS gold_daily_order_metrics (
  order_date DATE,
  order_month STRING,
  order_year INT,
  total_orders  INT,
  total_returns  INT,
  actual_orders  INT,
  net_revenue DECIMAL(18,2),
  total_revenue DECIMAL(18,2),
  total_return_amount DECIMAL(18,2),
  insert_timestamp TIMESTAMP,
  update_timestamp TIMESTAMP
)
```

```python
%sql
DROP TABLE IF EXISTS gold_profit_target_metrics;
CREATE TABLE IF NOT EXISTS gold_profit_target_metrics (
  month_of_order_date DATE,
  category STRING,
  target_profit DECIMAL(18,2),
  actual_profit DECIMAL(18,2),
  profit_gap DECIMAL(18,2),
  target_achieved_flag BOOLEAN,
  target_achievement_pct DECIMAL(5,2),
  insert_timestamp TIMESTAMP,
  update_timestamp TIMESTAMP
)
```
